## # Using KNN algo to detect non linear boundaries

```python
from numpy import *
import numpy as np
import pandas as pd
import operator
from os import listdir
```

```python
# normalization with min/max
def autoNorm(dataSet):
    minVals = dataSet.min(0)
    maxVals = dataSet.max(0)
    ranges = maxVals - minVals
    normDataSet = zeros(shape(dataSet))
    m = dataSet.shape[0]
    normDataSet = dataSet - tile(minVals, (m,1))
    normDataSet = normDataSet/tile(ranges, (m,1))  
    return normDataSet
```

```python
# normalization with std deviation
def autoNormWithStd(dataSet):
    dataSetMean = dataSet.mean(axis=0)
    dataSetStd = dataSet.std(axis=0)
    return (dataSet - dataSetMean)/dataSetStd
```

```python
def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
    diffMat = tile(inX, (dataSetSize,1)) - dataSet
    sqDiffMat = diffMat**2
    sqDistances = sqDiffMat.sum(axis=1)
    distances = sqDistances**0.5
    sortedDistIndices = distances.argsort()     
    classCount={}          
    for i in range(k):
        voteIlabel = labels[sortedDistIndices[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel,0) + 1
    sortedClassCount = sorted(classCount.iteritems(), key=operator.itemgetter(1), reverse=True)
    return sortedClassCount[0][0], sortedDistIndices[:k]
```

```python
def file2matrix(filename,cols,sep='\t'):
    fr = open(filename)
    numberOfLines = len(fr.readlines())       
    returnMat = zeros((numberOfLines,cols))       
    classLabelVector = []              
    fr = open(filename)
    index = 0
    for line in fr.readlines():
        line = line.strip()
        listFromLine = line.split(sep)
        returnMat[index,:] = listFromLine[0:cols]
        classLabelVector.append(int(listFromLine[-1]))
        index += 1
    return returnMat,classLabelVector
```

```python
def plot2d(data, ax=None, class0marker=('b','o'), class1marker=('y','x')):
  if (ax == None):
    fig=plt.figure(figsize=(10,8))
    ax1=fig.add_subplot(1,1,1)
  else:
    ax1 = ax
  class0Idx=data[:,2] == 0
  class1Idx=data[:,2] == 1
  ax1.scatter(data[class0Idx,0],data[class0Idx,1], c=class0marker[0], marker=class0marker[1])
  ax1.scatter(data[class1Idx,0],data[class1Idx,1], c=class1marker[0], marker=class1marker[1])
  return ax1
ax=plot2d(data)

#testData=data[:0.3 * data.shape[0],:]
#testData.shape
#plot2d(testData,ax,('r','o'),('r','x'))
```

```python
def courseraClassTest():
    # NOTE: error rate varies a lot with k. Maybe need more data (?)
    k=3
    hoRatio = 0.20      # percentage taken for the test set
    courseraDataMat,courseraLabels = file2matrix("ex2data2.txt", cols=2, sep=',')
    # shuffle rows since not not well shuffled in this data set: test set was biased
    rows=courseraDataMat.shape[0]
    sampler=np.random.permutation(rows)
    courseraDataMat = courseraDataMat[sampler][:]
    labels=np.array(courseraLabels)
    courseraLabels=labels[sampler].tolist()
    
    ax=plot2d(np.concatenate((courseraDataMat,array(courseraLabels).reshape(-1,1)),axis=1))
    
    #normMat = autoNorm(courseraDataMat) # norm with min/max
    normMat = autoNormWithStd(courseraDataMat) # norm with stddev
    # normMat = courseraDataMat # no norm
    m = normMat.shape[0]
    numTestVecs = int(m*hoRatio)
    errorCount = 0.0
    for i in range(numTestVecs):
      classifierResult,_ = classify0(normMat[i,:],normMat[numTestVecs:m,:],courseraLabels[numTestVecs:m],k)
      if (i < 0):
        print "the classifier came back with: %d, the real answer is: %d" % \
          (classifierResult, courseraLabels[i])
      if (classifierResult != courseraLabels[i]): errorCount += 1.0
    print "the total error rate is: %f" % (errorCount/float(numTestVecs))
    print "error count: %d" % errorCount

    # show the prediction for one random point and its k nearest neighbors
    ptIdx=0
    classifierResult,nnIndices=classify0(normMat[ptIdx,:],normMat[numTestVecs:m,:],courseraLabels[numTestVecs:m],k)
    ptSet=zeros((1,3)) # 1 row, 3 columns (2 features + 1 label)
    ptSet[0,0:2]=courseraDataMat[ptIdx,:]
    ptSet[0,2]=courseraLabels[ptIdx]
    print "point: %s" % str(ptSet)
    print "predicted label: %s" % classifierResult
    # show actual point in red with shape for actual label
    ax=plot2d(ptSet,ax,('r','o'),('r','x'))            
    # show k nearest neighbors in green
    nn=np.concatenate(((courseraDataMat[numTestVecs:m,:])[nnIndices,:],(array(courseraLabels)[numTestVecs:m])[nnIndices].reshape(-1,1)),axis=1)
    print nn
    ax=plot2d(nn,ax,('g','o'),('g','x'))
```
