```python
def plot2d(data, ax=None, class0marker=('b','o'), class1marker=('y','x')):
  if (ax == None):
    fig=plt.figure(figsize=(10,8))
    ax1=fig.add_subplot(1,1,1)
  else:
    ax1 = ax
  class0Idx=data[:,2] == 0
  class1Idx=data[:,2] == 1
  ax1.scatter(data[class0Idx,0],data[class0Idx,1], c=class0marker[0], marker=class0marker[1])
  ax1.scatter(data[class1Idx,0],data[class1Idx,1], c=class1marker[0], marker=class1marker[1])
  return ax1
```

```python
z=data[:,2:3]
zpos=(z == 1).ravel()
zneg=(z == 0).ravel()
posdata=data[zpos,:]
negdata=data[zneg,:]
xpos=posdata[:,0:1]
xneg=negdata[:,0:1]
ypos=posdata[:,1:2]
yneg=negdata[:,1:2]
plt.scatter(xpos, ypos,c='black',marker='+')
plt.scatter(xneg, yneg,c='y')

m=shape(data)[0]
features=shape(data)[1]-1 # num of columns minus the last one (that is the Y value)
originalX=data[:,0:features]
print "data:"
print data[0:5]
print "X before bias:"
print originalX[0:5]
X=np.concatenate((ones((m,1)),originalX),axis=1)
y=data[:,[features]]
print "X with biased term:"
print X[0:5]
print "y:"
print y[0:5]
initial_theta=zeros(features+1)
```

```python
def Jtheta(theta,X,y,costList=None):
  """ theta must be a ndarray of 1 dimension """
  # print "theta: " + str(theta)
  hthetaX=1/(1+exp(-X.dot(c_[theta])))
  m=shape(X)[0]
  cost= (-y*log(hthetaX) - (1-y)*log(1-hthetaX)).sum() / m
  if costList is not None:
    costList.append(cost)
    # print "cost: " + str(cost)
  return cost

def computeGrad(theta,X,y,costList=None):
  """ theta must be a ndarray of 1 dimension """
  # print "computeGrad: theta: " + str(theta)
  m=shape(X)[0]
  hthetaX=1/(1+exp(-X.dot(c_[theta])))
  errors=hthetaX-y
  dtheta=(errors * X).sum(axis=0) / m
  return dtheta

# to optimize theta with gradient descent manually (iteratively)
def newtheta(theta, X,y,alpha):
  """ theta must be a ndarray of 1 dimension """
  # return theta - alpha * (computeGrad(theta,X,y).reshape(-1,1))
  return theta - alpha * computeGrad(theta,X,y)

def findTheta(X, y, iterations, alpha, initial_theta=None, costList=None):
    """
    Returns a tuple containing:
    - the converged theta vector 
    - a vector containing the cost for each iteration.
    """
    features=shape(X)[1]
    if initial_theta is None:
        theta=zeros(features)
    else:
        theta = initial_theta
    for i in range(iterations):
        theta=newtheta(theta, X,y,alpha)
        if costList is not None:
          costList.append(Jtheta(theta,X,y))
    return theta,costList
```

```python
def plotCostHistory(costList):
    fig = plt.figure(figsize=(8,6))
    ax=fig.add_subplot(1,1, 1)
    ax.plot(costList)
    ax.set_title("Cost per iteration")
```

```python
# try with a learning rate of 0.001
theta,costList=findTheta(X,y,20000,alpha=
                         0.001,initial_theta=None,costList=[])
plotCostHistory(costList)
print "cost at last iter: " + str(Jtheta(theta,X,y,costList=None))
```

```python
# if we continue with the same learning rate, it would take millions of iterations
# to approach the minimum. So after 20000 iterations, increase alpha to 0.004:
# there will be some oscillations but it'll converge to the min value after 
# approx 300000 iterations. Play with the below code to change alpha and
# continue from the previous result repeatedly
theta,costList=findTheta(X,y,200000,alpha=0.004,initial_theta=theta,costList=costList)
plotCostHistory(costList)
print "cost at last iter: " + str(Jtheta(theta,X,y,costList=None))
# finally the min cost will be reached: approx 0.20367
```

## Regularized logistic regression
### Expanding features from original ones: quadratic combinations

```python
# map the two features into 28 features for quadratic combinations of 
# the two features
def map_feature(x1, x2):
  degree = 6
  colsNb = (degree+1)*(degree+2)/2
  rowsNb = x1.shape[0]
  X=zeros((rowsNb,colsNb))
  col=0
  for i in range(degree+1):
    for j in range(i+1):
      X[:,col] = x1**(i-j) * x2**(j)
      #print str(i-j) + "," + str(j)
      col += 1
  return X
```

```python
data=np.loadtxt("ex2data2.txt", delimiter=',')
data[:5]
```

```python
ax=plot2d(data)
```

```python
# prepare the X matrix (the 28 features) and the y vector containing the labels
x1=data[:,0]
x2=data[:,1]
y=c_[data[:,2]]
X=map_feature(x1,x2)
X.shape
```

```python
def Jtheta_reg(theta, X,y,lbda,costList=None):
  """ theta must be a ndarray of 1 dimension """
  hthetaX=1/(1+exp(-X.dot(c_[theta])))
  m=shape(X)[0]
  cost= (-y*log(hthetaX) - (1-y)*log(1-hthetaX)).sum() / m
    
  # regularization term
  coef=ones(shape(X)[1])
  coef[0]=0
  cost_reg = cost + lbda/(2.0*m)*((coef*theta)**2)[1:].sum()
  if costList is not None:
    costList.append(cost_reg)
  return cost_reg

def computeGrad_reg(theta, X,y,lbda,costList):
  """ theta must be a ndarray of 1 dimension """
  m=shape(X)[0]   
  hthetaX=1/(1.0+exp(-X.dot(c_[theta])))
  errors=hthetaX-y
  dtheta=(errors * X).sum(axis=0) / m
    
  # regularization term
  coef=ones(shape(X)[1])
  coef[0]=0
  dtheta_reg = dtheta + lbda / float(m) * (coef*theta)
  return dtheta_reg
```

```python
# use bfgs to find minimum
from scipy import optimize
options = {'full_output': True, 'maxiter': 50}
features=shape(X)[1]
initial_theta=zeros(X.shape[1])
lbda=1
costList=[]
theta, cost,_,_,funcalls,gradcalls,warnflag,gradhist = \
  optimize.fmin_bfgs(Jtheta_reg, initial_theta, fprime=computeGrad_reg, \
                     args=(X,y,lbda,costList), retall=1, **options)
print 'Cost at theta found by fminunc: %f' % cost
print 'theta: %s' % theta
print 'funcalls: %s' % funcalls
print 'gradcalls: %s' % gradcalls
print 'warnflag: %s' % warnflag

plotCostHistory(costList)
```

### Draw Grid of Predictions

```python
(test_x,test_y)=np.meshgrid(arange(-1,1.2, 0.05),arange(-1,1.2, 0.05))
test_arr=concatenate((c_[test_x.flatten()],c_[test_y.flatten()]),axis=1)
# compute the mapped original vector to the 28 features
mapped_arr=map_feature(test_arr[:,0],test_arr[:,1])
mapped_arr.shape
```

```python
def predictY(theta,X):
    hthetaX=1/(1+exp(-X.dot(c_[theta])))
    flattened=hthetaX.flatten()
    prediction=zeros(len(flattened))
    for i in range(0, X.shape[0]):
        if flattened[i]>=0.5:
            prediction[i] = 1
    return prediction

predicted_y=predictY(theta,mapped_arr)
test_arr_with_y=concatenate((test_arr,c_[predicted_y]), axis=1)
ax=plot2d(test_arr_with_y)
```

```python
def computeAccuracy(theta,X,y):
    hthetaX=1/(1+exp(-X.dot(c_[theta])))
    flattened=hthetaX.flatten()
    prediction=zeros(len(flattened))
    for i in range(0, len(y)):
        #print flattened[i]
        if flattened[i]>=0.5:
            prediction[i] = 1
    successCnt=0
    for i in range(0, len(y)):
        if prediction[i] == y[i]:
            successCnt += 1
    print "train accuracy: %f" % (successCnt / float(len(y)))

computeAccuracy(theta,X,y=c_[data[:,2]])
```
